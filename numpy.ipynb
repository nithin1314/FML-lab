**as array of C row**

```python
import numpy as nk
a=[200,300,400,500]
b=np.asarray(a,order="C")
print(b)
```

```python
import numpy as nk
a=[[200,300],[400,500]]
b=np.asarray(a,order="C")
print(b)
for i in nk.nditer(b):
  print(i)
```
